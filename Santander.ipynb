# An augmentation function (oversample)

```python
def augmentation(x,y,t=2):
    xs,xn = [],[]
    for i in range(t):
        mask = y>0
        x1 = x[mask].copy()
        ids = np.arange(x1.shape[0])
        for c in range(x1.shape[1]):
            np.random.shuffle(ids)
            x1[:,c] = x1[ids][:,c]
        xs.append(x1)

    for i in range(t//2):
        mask = y==0
        x1 = x[mask].copy()
        ids = np.arange(x1.shape[0])
        for c in range(x1.shape[1]):
            np.random.shuffle(ids)
            x1[:,c] = x1[ids][:,c]
        xn.append(x1)

    xs = np.vstack(xs); xn = np.vstack(xn)
    ys = np.ones(xs.shape[0]);yn = np.zeros(xn.shape[0])
    x = np.vstack([x,xs,xn]); y = np.concatenate([y,ys,yn])
    return x,y
```

# Training

```python
params = {
    'max_depth': -1,
    'metric':'auc',
    'min_data_in_leaf': 80,
    'bagging_freq': 5,
    'bagging_fraction': 0.335,
    'boost_from_average':'false',
    'boost': 'gbdt',
    'feature_fraction': 0.041,
    'learning_rate': 0.0083,
    'min_sum_hessian_in_leaf': 10.0,
    'num_leaves': 13,
    'num_threads': 8,
    'tree_learner': 'serial',
    'objective': 'binary', 
    'verbosity': -1,
    'n_jobs': -1,
    #'sparse_threshold':1.0,
    #'device' :'gpu',
    #'max_bin' : 63,
    #'gpu_use_dp':True
}
```

```python
%time
oof = train[['ID_code', 'target']]
oof['predict'] = 0
predictions = test[['ID_code']]
val_aucs = []
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 55)
for counter, (fit_index, valid_index) in enumerate(skf.split(train, train['target'])):
    print("Fold: {}".format(counter+1))
    #Train df
    X_fit, y_fit = train.iloc[fit_index][features], train.iloc[fit_index]['target']
    #Validation df
    X_val, y_val = train.iloc[valid_index][features],train.iloc[valid_index]['target']
    
    N =3
    p_valid, yp = 0, 0
    for i in range(N):
        X_f, y_f = augmentation(X_fit.values, y_fit.values)
        X_f = pd.DataFrame(X_f)
        fit_data = lgb.Dataset(X_f, label = y_f)
        valid_data = lgb.Dataset(X_val, label = y_val)
        
        model = lgb.train(params, fit_data, 100000, 
                     valid_sets = [fit_data, valid_data],
                     early_stopping_rounds = 4000,
                     verbose_eval = 1000)
        p_valid += model.predict(X_val)
        yp += model.predict(test[features])
    oof['predict'][valid_index] = p_valid/N
    val_score = roc_auc_score(y_val, p_valid)
    val_aucs.append(val_score)
    predictions['fold{}'.format(counter+1)] = yp/N
```

```python
print("CV score: {:<8.5f}".format(roc_auc_score(oof['target'], oof['predict'])))
```

```python
predictions['target'] = np.mean(predictions[[col for col in predictions.columns if col not in ['ID_code', 'target']]].values, axis=1)
sub_df = pd.DataFrame({"ID_code":test["ID_code"].values})
sub_df["target"] = predictions['target']
sub_df.to_csv("submission7.csv", index=False)
```
